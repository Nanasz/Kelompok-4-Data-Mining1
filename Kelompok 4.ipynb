```python
## Hapus missing values
data_rumah.dropna(inplace=True)
print(data_rumah.isnull().sum())
```

```python
data_rumah['hadap'].value_counts()
```

```python
#replace beberapa missing value
data_rumah['KM'] = data_rumah['KM'].replace('-','5')
data_rumah['carport'] = data_rumah['carport'].replace('-','0')
data_rumah['KT'] = data_rumah['KT'].replace('-','2')
data_rumah['listrik'] = data_rumah['listrik'].replace('-','1300')
```

```python
#mengecek duplicate
data_rumah.drop_duplicates(inplace=True)
data_rumah.duplicated().sum()
```

```python
missing_data = pd.DataFrame({'total_missing': data_rumah.isnull().sum(), 'perc_missing': (data_rumah.isnull().sum()/102)*100})
missing_data
```
